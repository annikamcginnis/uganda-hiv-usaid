```python
matched_rows = art_pop[art_pop['Subcounty Population'].notna()]
unmatched_rows = art_pop[art_pop['Subcounty Population'].isna()]
```

```python
#Create a version of merged_df with Subcounties that occur only once
unique_subcounties = merged_df['Subcounty'].value_counts()
single_entry_subcounties = unique_subcounties[unique_subcounties == 1].index
merged_df_single = merged_df[merged_df['Subcounty'].isin(single_entry_subcounties)]
```

```python
# Step 4: Try merging the unmatched rows again, this time only on 'Subcounty'
retry_merge = pd.merge(
    unmatched_rows.drop(columns=[col for col in merged_df.columns if col in unmatched_rows.columns and col != 'Subcounty']),
    merged_df_single,
    on='Subcounty',
    how='left'
)
```

```python
# Step 5: Combine the matched rows and re-merged unmatched rows
final_merged = pd.concat([matched_rows, retry_merge], ignore_index=True)
```

```python
from fuzzywuzzy import process
```

```python
def get_best_fuzzy_match(pirs_row, merged_df, threshold=97):
    subcounty = str(pirs_row['PIRS Subcounty']) if pd.notna(pirs_row['PIRS Subcounty']) else ""

    # Handle edge case where subcounty is empty
    if subcounty == "":
        return None, None

    # Get the best fuzzy match for the subcounty from the entire merged_df
    result = process.extractOne(subcounty, merged_df['Subcounty'].dropna().astype(str))

    if result:  # Check if result is not None
        best_match = result[0]
        score = result[1]
        
        if score >= threshold:
            matched_row = merged_df[merged_df['Subcounty'] == best_match].iloc[0]
            return matched_row['Subcounty Population'], best_match
    return None, None
```

```python
missing_rows[['Estimated Population', 'Best Match Subcounty']] = missing_rows.apply(
    lambda row: pd.Series(get_best_fuzzy_match(row, merged_df)), axis=1
)
```
